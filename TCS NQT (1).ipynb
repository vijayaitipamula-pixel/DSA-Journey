```python
#check even or odd
def even_or_odd(n):
    if n%2==0:
        print('Even')
    else:
        print('Odd')
n=int(input())
even_or_odd(n)
```

```python
#check prime or not
def prime_or_not(n):
    if n<=1:
        print('not prime')
    else:
        c=0
        for i in range(2,n//2+1):
            if n%i==0:
                c+=1
        if c>=1:
            print('not prime')
        else:
            print('prime')
n=int(input())
prime_or_not(n)
```

```python
#factorial of a number
def fact_num(n):
    fact=1
    for i in range(2,n+1):
        fact*=i
    print(fact)
n=int(input())
fact_num(n)
```

```python
#Fibonacci series(first n numbers)
def fibo(n):
    prev=0
    curr=1
    print(prev,end=',')
    print(curr,end=',')
    for i in range(n-2):
        new_num=prev+curr
        prev=curr
        curr=new_num
        print(new_num,end=',')
n=int(input())
fibo(n)
```

```python
#reverse a number
def reversing(n):
    rev=0
    for i in range(len(str(n))):
        last_dig=n%10
        rev=rev*10+last_dig
        n=n//10
    return rev
n=int(input())
print(reversing(n))
```

```python
#check polindrome number
def is_polindrome(n):
    rev=0
    temp=n
    for i in range(len(str(temp))):
        last_dig=temp%10
        rev=rev*10+last_dig
        temp=temp//10
    
    if rev==n:
        return True
    else:
        return False
n=int(input())
print(is_polindrome(n))  
```

```python
#armstrong number
def armstrong(n):
    length=len(str(n))
    temp=n
    s=0
    for i in range(length):
        last_dig=temp%10
        s+=last_dig**length
        temp=temp//10
    if n==s:
        return True
    else:
        return False
n=int(input())
print(armstrong(n))
```

```python
#sum of digits
def sum_of_dig(n):
    s=0
    temp=n
    for i in range(len(str(n))):
        last_dig=temp%10
        s=s+last_dig
    return s
n=int(input())
print(sum_of_dig(n))
```

```python
#longest sub String
def longest_sub_string(s):#s=pwwkew
    char_set=set()
    left=0
    max_len=0
    for right in range(len(s)):
        while s[right] in char_set:
            char_set.remove(s[left])
            left+=1
        char_set.add(s[right])
        max_len=max(max_len,right-left+1)
    return max_len
s=input().strip()
print(longest_sub_string(s))
    
```
